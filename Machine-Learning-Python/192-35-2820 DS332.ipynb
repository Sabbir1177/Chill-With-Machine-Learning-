# Import the required Libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Reading our Dataset

```python
df=pd.read_csv("Real estate.csv")
df
```

```python
df.head()
```

# Visualizing the Data using HeatMap

```python
sns.heatmap(df.isnull(),cbar=False,cmap="viridis")
```

# Check the correlations in the dataset

```python
df.corr()
```

# Make features

```python
x=df["X4 number of convenience stores"].values.reshape(-1,1)
y=df["Y house price of unit area"].values.reshape(-1,1)
y
```

```python
x
```

# Train test splitting

```python
from sklearn.model_selection import train_test_split
train_set_x,test_set_x,train_set_y,test_set_y=train_test_split(x,y,test_size=0.3,shuffle=True)
train_set_x.shape
```

# Generating the model and train regression model  

```python
from sklearn.linear_model import LinearRegression
r=LinearRegression()
r.fit(train_set_x,train_set_y)
```

```python
y_predict=r.predict(test_set_x)
y_predict
```

# Check Coefficients, Intercept, R-Squared Value

```python
r.coef_
```

```python
r.intercept_
```

```python
y_pred=r.intercept_+r.coef_+test_set_x
y_pred
```

# Evaluate the model and give a proper view of performance 

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
print("actual_data:",test_set_y ,"predictive values:",y_pred)
mean_absolute_error(test_set_y,y_pred),np.sqrt(mean_squared_error(test_set_y,y_pred))
```

```python
plt.scatter(test_set_x,test_set_y,color ="green")
plt.scatter(test_set_x,y_pred,color ="red")
```

```python
plt.scatter(train_set_x,train_set_y, color ="green" , alpha= 0.7)
plt.plot(train_set_x, r.predict(train_set_x), color = "red")
plt.xlabel('X4 number of convenience stores')
plt.ylabel('Y house price of unit area')
plt.show()
```
